```python
import pandas as pd
import os
from bs4 import BeautifulSoup as bs
```

```python
## Create a BeautifulSoup object from the character usage data page

# Read html from saved usage data page
filepath = os.path.join("Resources", "usagestats.html")
with open(filepath, encoding="utf-8") as file:
    html = file.read()

# Create object
soup = bs(html, 'html.parser')
soup
```

```python
# Isolate the data table

results = soup.find_all("div", class_="ratioBox")
results

# Get a list of the corresponding usage percentages

#percent_results = soup.find_all("p", class_="percent")

#print(character_results)
#print(percent_results)
```

```python
# Create empty lists to store character names and percentages
characters = []
percents = []

# Loop through result to populate lists
for result in results:
    
    # Get character names
    character = result.find("p", class_="name")
    character_text = character.text
    print(character_text)
    characters.append(character_text)
    
    # Get percentages
    percent = result.find("p", class_="percent")
    percent_text = percent.text
    print(percent_text)
    percents.append(percent_text)
    
print(characters)
print(percents)
```

```python
# Loop through percent list to convert strings to floats
for percent in percents:
    float(percent.strip("%"))

# Zip lists into a dictionary

usage_percents = dict(zip(characters, percents))
print(usage_percents)
```

```python
usage_df = pd.Series(usage_percents).to_frame("percentage")
usage_df.index.name = "character"
usage_df.head()
```

```python
# Reset index so character names are in a column
usage_df = usage_df.reset_index()
usage_df.head()
```

```python
usage_df.dtypes

# Convert percent strings to float
usage_df["percentage"].str.rstrip("%").astype("float")
usage_df.head()
```
