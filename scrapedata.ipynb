```python
import pandas as pd
import os
from bs4 import BeautifulSoup as bs
```

```python
## Create a BeautifulSoup object from the character usage data page

# Read html from saved usage data page
filepath = os.path.join("Resources", "usagestats.html")
with open(filepath, encoding="utf-8") as file:
    html = file.read()

# Create object
soup = bs(html, 'html.parser')
soup
```

```python
# Isolate the data table

results = soup.find_all("li", class_="big")
results
```

```python
# Create empty lists to store character names, percentages, and links for character images
characters = []
percents = []
image_urls = []

# Loop through result to populate lists
for result in results:
    
    # Get character names
    character = result.find("p", class_="name")
    character_text = character.text
    print(character_text)
    characters.append(character_text)
    
    # Get percentages
    percent = result.find("p", class_="percent")
    percent_text = percent.text
    print(percent_text)
    percents.append(percent_text)
    
    # Get url for character image
    image = result.find("img")
    url = image["src"]
    print(url)
    image_urls.append(url)
    
print(characters)
print(percents)
print(image_urls)
```

```python
# Repeat last two code blocks for li elements with class "small"

results = soup.find_all("li", class_="small")
results
```

```python
# Loop through result to populate lists
for result in results:
    
    # Get character names
    character = result.find("p", class_="name")
    character_text = character.text
    print(character_text)
    characters.append(character_text)
    
    # Get percentages
    percent = result.find("p", class_="percent")
    percent_text = percent.text
    print(percent_text)
    percents.append(percent_text)
    
    # Get url for character image
    image = result.find("img")
    url = image["src"]
    print(url)
    image_urls.append(url)
    
    
print(characters)
print(percents)
print(image_urls)
```

```python
# Create dataframe using the three lists as columns

usage_df = pd.DataFrame({"character": characters, "usage_percentage": percents, "image_url": image_urls})
usage_df.head()
```

```python
# Convert percent strings to float
usage_df["usage_percentage"] = usage_df["usage_percentage"].str.rstrip("%").astype("float")
usage_df.head()
```

```python
## Create a BeautifulSoup object from the character usage data page

# Read html from saved usage data page
filepath = os.path.join("Resources", "combatstats.html")
with open(filepath, encoding="utf-8") as file:
    html = file.read()

# Create object
soup = bs(html, 'html.parser')
soup
```

```python
# Isolate data table

table = soup.find("div", class_="diaTable")
table
```

```python
# Find all table header html elements
headers = table.find_all("th")

# Delete first item in list, which does not have a link
del headers[0] # Use try/except?

# Create empty list to store links
header_image_urls = []

for header in headers:
    image = header.find("img")
    link = image["src"]
    header_image_urls.append(link)
    
print(header_image_urls)
```

```python
# Use pandas to read combat stats html table

df = pd.read_html("Resources/combatstats.html")
combat_df = df[0]
combat_df.head()
```

```python
# Add column for header_image_urls
combat_df.insert(0, "Selected Character", header_image_urls)
combat_df.head()
```

```python
# Rename Average column
combat_df = combat_df.rename(columns={combat_df.columns[1]: "Overall Win Rate"})
combat_df.head()
```

```python
# Add first two column titles to list to easily rename column headings
# Use try/except?

header_image_urls.insert(0, "Overall Win Rate")
header_image_urls.insert(0, "Selected Character")

combat_df.columns = header_image_urls
combat_df.head()
```

```python
# Create empty list to store character names to rename columns and change Selected Character values
character_names = []

# Create loop using df.loc to find the character name where the image urls match on both dataframes

for row in range(0, len(combat_df)):
    name = usage_df.loc[usage_df["image_url"] == combat_df["Selected Character"][row], ["character"]].values[0][0]
    character_names.append(name)
    
print(character_names)
```

```python
# Change values in Selected Character column to character names

combat_df["Selected Character"] = character_names
combat_df.head()
```

```python
# Store old column names in a list
old_cols = combat_df.columns

# Set new_cols variable equal to character names list
new_cols = character_names

# Insert first two column names into new_cols so both lists have an equal amount of items
new_cols.insert(0, old_cols[1])
new_cols.insert(0, old_cols[0])
print(new_cols)
```

```python
# Rename columns in combat_df by zipping old and new column lists into a dictionary
combat_df.rename(columns = dict(zip(old_cols, new_cols)), inplace=True)
combat_df.head()
```
